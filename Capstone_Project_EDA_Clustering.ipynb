```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
def treat_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for feature in improved_df.iloc[:,0:14].columns: 
    lr,ur=treat_outlier(improved_df[feature])
    improved_df[feature]=np.where(improved_df[feature]>ur,ur,improved_df[feature])
    improved_df[feature]=np.where(improved_df[feature]<lr,lr,improved_df[feature])
```
